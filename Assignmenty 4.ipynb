```python
# IMPORT NUMPY and PANDAS

import numpy as np
import pandas as pd
```

```python
# QUESTION 2

# I decided to choose the 2021-22 NBA Season for my league and year
# I chose the 2022 Warriors, Celtics, Mavericks, and Heat (the 4 conference finals teams)
# I currently have 4 separate CSVs saved (1 per roster)
```

```python
# QUESTION 3

# Now I am going to pull in all 4 CSVs for each roster that I chose
# I also want to have player name be the index

warriors_roster = pd.read_csv("Golden State Warriors Roster 2021-22.csv",index_col='Player')
celtics_roster = pd.read_csv("Boston Celtics Roster 2021-22.csv",index_col='Player')
mavericks_roster = pd.read_csv("Dallas Mavericks Roster 2021-22.csv",index_col='Player')
heat_roster = pd.read_csv("Miami Heat Roster 2021-22.csv",index_col='Player')
```

```python
# Knowing that I end up combining these into one DataFrame, I am going to add Team as a column
# I know this was not part of the assignment, but this is more for overall practice and data clarity

warriors_roster['Team'] = 'Warriors'
celtics_roster['Team'] = 'Celtics'
mavericks_roster['Team'] = 'Mavericks'
heat_roster['Team'] = 'Heat'
```

```python
# QUESTION 4

# Now that I have all 4 rosters read in and stored into their own DataFrame, I can combine them into one

combined_rosters = pd.concat([warriors_roster,celtics_roster,mavericks_roster,heat_roster])
```

```python
# QUESTION 5

# Now that I have a new combined DataFrame containing all players from all rosters, I will write it to CSV

combined_rosters.to_csv("Combined Teams Roster 2021-22 - Warriors, Celtics, Mavericks, and Heat.csv")
```
